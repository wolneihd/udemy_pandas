```python
import matplotlib.pyplot as plt
```

```python
lista_temperaturas_fahrenheit = []
for temperatura in range(-50,51):
    lista_temperaturas_fahrenheit.append((temperatura * 1.8) + 32)
```

```python
# Criando o gráfico com duas linhas
plt.plot(range(-50,51), lista_temperaturas_fahrenheit, label='celcius', color='b')  

# Adicionando título e rótulos aos eixos
plt.title('Gráfico temperatura')
plt.xlabel('Celcius')
plt.ylabel('Fahrenheit')

# Adicionando grades internas
plt.grid(True, which='both', axis='both', linestyle='--', color='gray', alpha=0.5)

# Exibindo o gráfico
plt.show()
```
